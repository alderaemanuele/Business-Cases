```python
def f(x):
    if (x > 4) and (x <= 8):
        return 'Early Morning'
    elif (x > 8) and (x <= 12 ):
        return 'Morning'
    elif (x > 12) and (x <= 16):
        return'Noon'
    elif (x > 16) and (x <= 20) :
        return 'Eve'
    elif (x > 20) and (x <= 24):
        return'Night'
    elif (x <= 4):
        return'Late Night'

prods = pd.DataFrame({'hour':range(1, 25)})

b = [0,4,8,12,16,20,24]
l = ['Late Night', 'Early Morning','Morning','Noon','Eve','Night']
prods['session'] = pd.cut(prods['hour'], bins=b, labels=l, include_lowest=True)

data["DayTime"] = data['Hour'].apply(f)
```

```python
abc_transactions = abc_analysis(data["TotalPrice"], True, fltSize=10)

# index position of A, B, and C for transactions
transaction_a_index = abc_transactions['Aind']
transaction_b_index = abc_transactions['Bind']
transaction_c_index = abc_transactions['Cind']

# New Column indicating A, B, or C
condition = [data.index.isin(transaction_a_index),
             data.index.isin(transaction_b_index),
             data.index.isin(transaction_c_index)]

segment_list = ['A','B','C']

data['Transaction_ABC'] = np.select(condition, segment_list)
data.sort_values(by=['TotalPrice'], ascending=True)
abc_transactions
```

```python
customer_values = data[["CustomerID", "TotalPrice"]].groupby("CustomerID").sum("TotalPrice").sort_values("TotalPrice", ascending = False)
abc_customers = abc_analysis(customer_values["TotalPrice"], True, fltSize=10)

# index position of A, B, and C for transactions
customer_a_index = customer_values.iloc[abc_customers['Aind']].index
customer_b_index = customer_values.iloc[abc_customers['Bind']].index
customer_c_index = customer_values.iloc[abc_customers['Cind']].index

# New Column indicating A, B, or C
condition = [data["CustomerID"].isin(customer_a_index),
             data["CustomerID"].isin(customer_b_index),
             data["CustomerID"].isin(customer_c_index)]

segment_list = ['A','B','C']

data['Customer_ABC'] = np.select(condition, segment_list)
data.sort_values(by=['TotalPrice'], ascending=True)

# BEWARE WE HAVE SOME NaNs for these values --> might not be the most accurate classification
```

## Recommendation based on CustomerID

```python
def customer_recommender(customerid):
    
    
    # storing the last 3 bought items of the customer in a list
    l = data[data["CustomerID"]==customerid].iloc[-3:]["StockCode"].values.tolist()
    
    print("Based on the last 3 items of customer",customerid,":")
    for x in l:
        print(df_dict[x])
    
    print("\n","the following recommendations can be given:")
    # empty list to store the recommended Stockcodes in
    l_recommendations = []
    
    # if only one invoice for customer use the item recommender based on other users behavior
    
    if len(set(data[data["CustomerID"]==customerid].InvoiceNo.values)) <=1:
        for stockcode in l:
            l_items = item_recommender(stockcode,n=4)
            for item in l_items:
                l_recommendations.append(item)
    
    # if more transactions use a customized recommender
    else:
    
        # creating a dataframe with the customer's transactions
        df = data[data["CustomerID"]==customerid]
        df = df[["InvoiceNo","StockCode"]]
        df["flag"] = 1
    
        # Pivot table indicating the relationship between Invoice No and StockCode
        price_utility_matrix = df.pivot_table(index='InvoiceNo', columns='StockCode', fill_value=0)
    
        X = price_utility_matrix.T
        X.reset_index(inplace=True)
        X.set_index("StockCode",inplace=True)
        X.drop("level_0",axis=1,inplace=True)
    
        # Decomposing the Matrix if too big
        if len(price_utility_matrix) > 1:
    
            if len(price_utility_matrix) <= 10:
                n = len(price_utility_matrix) - 1
    
            else:
                n = 10
    
            SVD = TruncatedSVD(n_components=n)
            decomposed_matrix = SVD.fit_transform(X)
    
        else:
            decomposed_matrix = price_utility_matrix
    
        # Calculating the corrcoef
        correlation_matrix = np.corrcoef(decomposed_matrix)
    
    
    
        # looking for recommendations for each of the last 3 items
        for stockcode in l:
        
            if stockcode in list(X.index):
                l_description = []
                stockcode = str(stockcode)
                product_names = list(X.index)
                product_ID = product_names.index(stockcode)
                correlation_product_ID = correlation_matrix[product_ID]
                Recommend = list(X.index[correlation_product_ID > 0.90])

                # Removes the item already bought by the customer
                Recommend.remove(stockcode) 
                Recommend = Recommend[0:2]

                #         for j in Recommend:
                #             x = data[data["StockCode"]==j]["Description"].iloc[0]
                #             l_description.append(x)
            
            for recommendation in Recommend:
                l_recommendations.append(recommendation)
            
            # adding suggestions from other customers
            l_items = item_recommender(stockcode,n=2)
            for item in l_items:
                l_recommendations.append(item)

    # removing the None values
    try:
        while True:
            l_recommendations.remove(None)
    except ValueError:
        pass    

    # keeping only unique values
    l_rec = list(set(l_recommendations))
    for i in l_rec:
        print(df_dict[i])
        
    return l_rec
```

### Split

```python
def split_data(data):
    '''
    Splits dataset into training and test set.
    
    Args:
        data (pandas.DataFrame)
        
    Returns
        train_data (tc.SFrame)
        test_data (tc.SFrame)
    '''
    train, test = train_test_split(data, test_size = .2)
    train_data = tc.SFrame(train)
    test_data = tc.SFrame(test)
    return train_data, test_data
```
